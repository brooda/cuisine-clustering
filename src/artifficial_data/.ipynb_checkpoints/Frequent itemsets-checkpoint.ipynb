```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import scipy.spatial
from efficient_apriori import apriori
import pandas as pd

def openDataSet(data_folder):
    with open(f'../../data/artificial_data/{data_folder}/bought_products.p', 'rb') as filehandle:
        products = pickle.load(filehandle)

    with open(f'../../data/artificial_data/{data_folder}/dominant_cusine_for_user.p', 'rb') as filehandle:
        cuisine = pickle.load(filehandle)

    with open(f'../../data/artificial_data/{data_folder}/recipes_of_cuisine.p', 'rb') as filehandle:
        recipes = pickle.load(filehandle)

    return products, cuisine, recipes
```

```python
def itemsets_method():
    cuisine_to_itemsets = {}

    for cuisine in range(len(recipes_of_cuisine)):
        rec = recipes_of_cuisine[cuisine]
        itemsets, _ = apriori(rec, min_support=0.2,  min_confidence=1)

        cuisine_to_itemsets[cuisine] = itemsets

    ones = {}
    doubles = {}
    triples = {}
    quartets = {}


    for cuisine in range(len(recipes_of_cuisine)):
        dict_of_cuisine = cuisine_to_itemsets[cuisine]

        ones_list = []
        doubles_list = []
        triples_list = []
        quartets_list = []

        if 1 in dict_of_cuisine.keys():
            ones_list = list(dict_of_cuisine[1].keys())

        if 2 in dict_of_cuisine.keys():
            doubles_list = list(dict_of_cuisine[2].keys())

        if 3 in dict_of_cuisine.keys():
            triples_list = list(dict_of_cuisine[3].keys())

        if 4 in dict_of_cuisine.keys():
            quartets_list = list(dict_of_cuisine[4].keys())

        ones[cuisine] = ones_list
        doubles[cuisine] = doubles_list    
        triples[cuisine] = triples_list    
        quartets[cuisine] = quartets_list

    for cuisine in range(len(recipes_of_cuisine)):
        for other_cuisine in range(len(recipes_of_cuisine)):
            if cuisine != other_cuisine:
                ones[cuisine] = filter(lambda el: el not in ones[other_cuisine], ones[cuisine])
                ones[cuisine] = list(ones[cuisine])

                doubles[cuisine] = filter(lambda el: el not in doubles[other_cuisine], doubles[cuisine])
                doubles[cuisine] = list(doubles[cuisine])

                triples[cuisine] = filter(lambda el: el not in triples[other_cuisine], triples[cuisine])
                triples[cuisine] = list(triples[cuisine])

                quartets[cuisine] = filter(lambda el: el not in quartets[other_cuisine], quartets[cuisine])
                quartets[cuisine] = list(quartets[cuisine])

    perfect_match_rate = 0
    approximate_match_rate = 0
    counter = 0

    for user in range(1000):
        products = bought_products[user]
        true_cuisine = dominant_cusine_for_user[user]

        prox = np.zeros(20)


        for cuisine in range(20):
            on = ones[cuisine]
            do = doubles[cuisine]
            tr = triples[cuisine]
            qu = quartets[cuisine]

            rate = 0

            for el in on:
                elem = list(el)
                if all(x in products for x in elem):
                    rate += 1
            for el in do:
                elem = list(el)
                if all(x in products for x in elem):
                    rate += 2

            for el in tr:
                elem = list(el)
                if all(x in products for x in elem):
                    rate += 4

            for el in qu:
                elem = list(el)
                if all(x in products for x in elem):
                    rate += 8

            prox[cuisine] = rate


        if any(prox > 20):
            counter +=1

            top_cuisines = np.argsort(-prox)[:3]


            if top_cuisines[0] == true_cuisine:
                perfect_match_rate += 1

            if true_cuisine in top_cuisines:
                approximate_match_rate += 1

    return counter, perfect_match_rate/counter, approximate_match_rate/counter   
```

```python
data_folders = ["3 recipes from preferred cuisine", \
                "3 recipes from preferred cuisine not full", \
                "2 from preferred cuisine 1 from other", "3 from preferred cuisine 2 from other", \
                "2 from preferred cuisine 1 from other not full", \
                "3 from preferred cuisine 2 from other not full", \
                "3 from preferred cuisine 6 from others not full"]


for folder in data_folders:
    #print(folder)
    bought_products, dominant_cusine_for_user, recipes_of_cuisine = openDataSet(folder)
    counter, perfect_match_rate, approximate_match_rate = itemsets_method()
    
    print(np.round(perfect_match_rate, 3), np.round(approximate_match_rate, 3))
```
